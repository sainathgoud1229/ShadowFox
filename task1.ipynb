## 1. MATPLOTLIB
### Overview
Matplotlib is a 2D plotting library introduced by John D. Hunter in 2003.
Built on NumPy arrays. Foundation for many visualization libraries.
- Use cases: Scientific plots, academic reports, basic data visualization
- Key feature: Full control over every element of a plot

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 15, 30, 25])

plt.figure(figsize=(6, 4))
plt.plot(x, y, color='blue', linewidth=2, marker='o')
plt.title("Line Chart - Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Use case: Showing trends over time")
```

```python
categories = ['Python', 'Java', 'C++', 'JavaScript', 'R']
values = [85, 70, 60, 75, 50]

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='steelblue', edgecolor='black', width=0.5)
plt.title("Bar Chart - Programming Language Popularity")
plt.xlabel("Language")
plt.ylabel("Popularity Score")
plt.tight_layout()
plt.show()
print("Use case: Comparing categories")
```

```python
data = np.random.normal(loc=60, scale=15, size=200)

plt.figure(figsize=(6, 4))
plt.hist(data, bins=15, color='orange', edgecolor='black')
plt.title("Histogram - Student Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
print("Use case: Frequency distribution of data")
```

```python
x = np.random.rand(50) * 100
y = x * 0.8 + np.random.rand(50) * 20

plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='red', alpha=0.6, edgecolors='black')
plt.title("Scatter Plot - Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()
print("Use case: Finding correlation between two variables")
```

```python
labels = ['Python', 'Java', 'C++', 'Others']
sizes = [45, 25, 20, 10]
colors = ['#66b3ff', '#ff9999', '#99ff99', '#ffcc99']

plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Language Usage Share")
plt.tight_layout()
plt.show()
print("Use case: Showing percentage/proportion of categories")
```

## 2. SEABORN
### Overview
Seaborn is a high-level statistical visualization library built on Matplotlib.
Created by Michael Waskom. Produces beautiful plots with minimal code.
- Use cases: Statistical analysis, EDA, correlation analysis
- Key feature: Works directly with Pandas DataFrames

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
print("Datasets loaded!")
print(tips.head())
```

```python
plt.figure(figsize=(6, 4))
corr = tips.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap - Correlation Matrix")
plt.tight_layout()
plt.show()
print("Use case: Showing correlation between numerical variables")
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='day', y='total_bill', data=tips,
            hue='day', palette='Set2', legend=False)
plt.title("Box Plot - Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.tight_layout()
plt.show()
print("Use case: Showing data distribution and outliers")
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='day', y='tip', data=tips,
               hue='day', palette='muted', legend=False)
plt.title("Violin Plot - Tip Distribution by Day")
plt.tight_layout()
plt.show()
print("Use case: Full distribution shape of data")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='day', data=tips, hue='day', palette='viridis', legend=False)
plt.title("Count Plot - Number of Customers per Day")
plt.tight_layout()
plt.show()
print("Use case: Count frequency of categorical values")
```

```python
sns.pairplot(iris, hue='species', palette='Set1')
plt.suptitle("Pair Plot - Iris Dataset", y=1.02)
plt.show()
print("Use case: Relationships between all numerical columns")
```

## 3. COMPARISON: Matplotlib vs Seaborn

| Feature           | Matplotlib         | Seaborn                     |
|-------------------|--------------------|-----------------------------|
| Ease of Use       | Medium             | Easy                        |
| Customization     | Very High          | Medium                      |
| Interactivity     | Static             | Static                      |
| Statistical Plots | Manual             | Built-in                    |
| Best For          | Full control plots | Quick statistical EDA       |

### Conclusion
- Use Matplotlib when you need full control
- Use Seaborn for fast statistical visualization
- Both work best when used together
